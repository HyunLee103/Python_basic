```python
def check_winner(name,lst):
    if name in lst:
        return f'{name} 수상했다.'
    else:
        return f'{name} 수상하지 못했다.'
```

```python
winner=['향단이', '정민호', '김철수', '이영희', '홍길동']
```

```python
check_winner('김수지',winner)
```

```python
check_winner('김철수',winner)
```

```python
check_winner('향단이',winner)
```

```python
check_winner('성춘향',winner)
```
